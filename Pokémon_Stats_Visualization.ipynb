### Pokemon Stats Visulization

Downloaded the file [Pokemon.csv](https://drive.google.com/file/d/1STSYKp8Zo8hJ0svOpKJXrwaKv41ZO7hw/view?usp=sharing) and read it into a pandas dataframe.  



```python
import pandas as pd
df= pd.read_csv('Pokemon.csv')
df.head()
```

Droped the column labeled `#`

```python
df.drop(['#'],axis=1,inplace=True)
df
```

Created a scatterplot of `Attack` vs. `Defense`, where the color of a point is determined by the `Total` column. Use `viridis` for the colormap.

```python
df.plot.scatter(x='Attack',y='Defense', colormap='viridis')
```

Created a scatterplot matrix of the HP, Attack, Defense, Sp. Atk, Sp. Def and Speed columns.

```python
splom=pd.plotting.scatter_matrix(df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']])
```

Generated a list of unique values for the Type 1 column.

```python
type=df['Type 1'].unique()
print(type)
```

Generated a list of unique values for the `Type 2` column.

```python
df['Type 2'].unique()
```

A bar chart that shows how many Pokemon there are for each type given in `Type 1`.

```python
df.groupby(['Type 1']).plot.bar(x='Type 1')
```

 Generated a correlation matrix for the data-set, but only for the HP, Attack, Defense, Sp. Atk, Sp. Def and Speed columns.

```python
data= pd.DataFrame(df,columns=['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed'])
corrMatrix = data.corr()
print(corrMatrix)
```

Created a heatmap of the correlations from the previous exercise using the `RdYlGn` colormap.

```python
import seaborn as sb
sb.heatmap(df.corr(),cmap='RdYlGn')
```
